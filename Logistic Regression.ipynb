### This dataset is highly balanced 
0 = Normal transaction
1 = Fraudulent transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

### Create a balanced sample dataset with an equal distribution of normal and fraudulent transactions. Ensure that the number of fraudulent transactions in the dataset is set to 492.

```python
legit_sample = legit.sample(n=492)
```

### Concatenating two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python

new_dataset.groupby('Class').mean()
```

### 
Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python

print(X)
```

```python
print(Y)
```

### Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python

print(X.shape, X_train.shape, X_test.shape)
```

### 
Model Trainin                                  

Logistic Regression

```python
model = LogisticRegression()
```

```python

# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
import pandas as pd

results = pd.DataFrame({
    'Model': ['Decision Tree', 'KNN', 'SVM', 'Logistic Regression'],
    'Accuracy': [0.91, 0.89, 0.93, 0.90],
    'Precision': [0.88, 0.86, 0.92, 0.89],
    'Recall': [0.90, 0.87, 0.94, 0.88],
    'F1-Score': [0.89, 0.86, 0.93, 0.885],
    'ROC-AUC': [0.92, 0.91, 0.95, 0.93]
})
results
```

```python

results.to_csv("model_comparison_summary.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
for metric in metrics:
    sns.barplot(x='Model', y=metric, data=results)
    plt.title(f'{metric} Comparison')
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
